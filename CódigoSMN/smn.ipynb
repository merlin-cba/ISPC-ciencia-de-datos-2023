```python
import csv
from datetime import datetime
import pandas as pd
```

```python
def getData():
    with open('smn.csv', encoding='latin-1') as f:
        reader = csv.reader(f)
        salida = []
        first = True
        for row in reader:
            row = row[0].split(' ')
            if len(row[0]) == 7:
                row[0] = '0' + row[0]
                
            d = [*row[0:3] + [' '.join(row[3:-1]).strip()]]
            if first:
                first = False
                salida.append(d)
            else:
                try:
                    d = [datetime.strptime(d[0],'%d%m%Y').isoformat() ] + d[1:]
                except:
                    print(d)
                    continue
                if "CORDOBA" in d[3]:
                    salida.append(d)
        f.close()
    return salida
```

```python
salida = getData()
dataframe = pd.DataFrame(salida[1:], columns=salida[0]) 
dataframe
```

```python
dataframe.to_csv('cordoba.csv', sep=';')
```
